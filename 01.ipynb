# 1.데이터셋 불러오기

```python
import zipfile
```

```python
source_file = 'data/Train.zip'

with zipfile.ZipFile(source_file, 'r') as zf:
	zipinfo = zf.infolist()
	for info in zipinfo:
		info.filename = info.filename.encode('cp437').decode('euc-kr')
		zf.extract(info, 'data/Train')
```
